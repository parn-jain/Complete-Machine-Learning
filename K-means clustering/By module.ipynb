```python
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples = 500,n_features = 2,centers = 3,random_state = 23)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters = i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
wcss
```

```python
plt.plot(wcss)
```

```python
# it shows no of clusters should be 2
# 2 is the elbo point and this curve is the albow curve
```

```python
X = x[:,:]
km = KMeans(n_clusters = 3)
y_means = km.fit_predict(X)
y_means
```

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],color = 'green')
plt.scatter(X[y_means==1,0],X[y_means==1,1],color = 'blue')
plt.scatter(X[y_means==2,0],X[y_means==2,1],color = 'red')
# plt.scatter(X[y_means==3,0],X[y_means==3,1],color = 'blue')
 
```
